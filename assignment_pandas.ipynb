**4-5. Рассчитайте среднее значение и среднеквадратичное отклонение возраста тех, кто получает более 50K в год (признак *salary*) и тех, кто получает менее 50K в год**

**4-5. Salary-Based Age Statistics Calculate the mean and standard deviation of age for those who earn more than 50K per year (feature salary) and those who earn less than 50K per year.**

```python
mean_age_salary_more_than_50k = data[data['salary'].str.strip() == '>50K']['age'].mean()
std_salary_more_than_50k = data[data['salary'].str.strip() == '>50K']['age'].std()
print("Mean Age for those earning more than 50k:", mean_age_salary_more_than_50k)
print("Standard Deviation for those earning more than 50k:", std_salary_more_than_50k)

mean_age_salary_less_than_50k = data[data['salary'].str.strip() == '<=50K']['age'].mean()
std_salary_less_than_50k = data[data['salary'].str.strip() == '<=50K']['age'].std()
print("Mean Age for those earning more than 50k:", mean_age_salary_less_than_50k)
print("Standard Deviation for those earning more than 50k:", std_salary_less_than_50k)
```

**6. Правда ли, что люди, которые получают больше 50k, имеют минимум высшее образование? (признак *education – Bachelors, Prof-school, Assoc-acdm, Assoc-voc, Masters* или *Doctorate*)**

**6. Education Level vs. Salary Is it true that people who earn more than 50K all have at least a higher education? (feature education – Bachelors, Prof-school, Assoc-acdm, Assoc-voc, Masters, or Doctorate)**

```python
high_edu = ['Bachelors', 'Prof-school', 'Assoc-acdm', 'Assoc-voc', 'Masters', 'Doctorate']
salary_more_than_50k_high_edu = data[
    (data['salary'].str.strip() == '>50K') &
    (data['education'].str.strip().isin(high_edu))
]['education'].value_counts()
salary_more_than_50k_high_edu
```

```python
high_edu = ['Bachelors', 'Prof-school', 'Assoc-acdm', 'Assoc-voc', 'Masters', 'Doctorate']
salary_more_than_50k_not_high_edu = data[
    (data['salary'].str.strip() == '>50K') &
    (~data['education'].str.strip().isin(high_edu))
]['education'].value_counts()
salary_more_than_50k_not_high_edu
```

As you can see from the query, it is not true that every person earning more than 50k have at least higher education(Bachelors, Prof-school, Assoc-acdm, Assoc-voc, Masters, or Doctorate)

**Проставьте название соответсвтуещей группы для каждого человека в новой колонке AgeGroup**

**11.Age Grouping Group people into age groups: young, adult, and retiree, where:**
* *young* 16–35 years old
* *adult* 35–70 years old
* *retiree* 70–100 years old

**Assign the corresponding group name to each person in a new column named AgeGroup.**

```python
bins = [16, 35, 70, 100]
group_names = ['young', 'adult', 'retiree']
data['AgeGroup'] = pd.cut(data['age'], bins=bins, labels=group_names, include_lowest=True)
```

**14. Сгруппируйте людей по типу занятости (колонка occupation) и определите количество людей в каждой группе. После чего напишите функциюю фильтрации filter_func, которая будет возвращать только те группы, в которых средний возраст (колонка age) не больше 40 и в которых все работники отрабатывают более 5 часов в неделю (колонка hours-per-week)**

**14. Occupation Filtering Group people by occupation (column occupation) and determine the number of people in each group. Then, write a filtering function filter_func that returns only those groups where the average age is no more than 40 and where all workers work more than 5 hours per week.**

```python
occupation_group = data['occupation'].value_counts()
occupation_group
```

```python
def filter_func(x):
    cond1 = x['age'].mean() <= 40
    cond2 = (x['hours-per-week'] > 5).all()

    return cond1 and cond2
filtered_data = data.groupby('occupation').filter(filter_func)
result = filtered_data['occupation'].value_counts()
result
```
