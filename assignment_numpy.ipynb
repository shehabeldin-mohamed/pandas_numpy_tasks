**1.** Дан случайный массив, поменять знак у элементов, значения которых между 3 и 8

**1.** Given a random array, change the sign of elements with values between 3 and 8.

```python
import random

rand_arr = np.random.randint(0,10, size=10)
print("Original Array:", rand_arr)

rand_arr[(rand_arr >=3) & (rand_arr <=8)] = -rand_arr[(rand_arr >=3) & (rand_arr <=8)]
print("Modified Array:", rand_arr)
```

**2.** Заменить максимальный элемент случайного массива на 0

**2.** Replace the maximum element of a random array with 0.

```python
import random

rand_arr = np.random.randint(0,10, size=10)
print("Original Array:", rand_arr)

rand_arr[rand_arr == np.max(rand_arr)] = 0
print("Modified Array:", rand_arr)

```
