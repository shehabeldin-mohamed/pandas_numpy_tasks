**3.** Построить прямое произведение массивов (все комбинации с каждым элементом). На вход подается двумерный массив

**3.** Construct the Cartesian product of arrays (all combinations of every element). The input is a two-dimensional array.

```python
def cartesian_product(arr):
  grid = np.meshgrid(*arr,indexing='ij')
  return np.stack(grid, axis=-1).reshape(-1, len(arr))

rand_arr = np.random.randint(0,10, size=(2,4))
print("Original Array:", rand_arr)

print("Cartesian Product:", cartesian_product(rand_arr))
```

**4.** Даны 2 массива A (8x3) и B (2x2). Найти строки в A, которые содержат элементы из каждой строки в B, независимо от порядка элементов в B

**4.** Given two arrays A (8x3) and B (2x2). Find the rows in A that contain elements from every row in B, regardless of the order of elements in B

```python
A = np.random.randint(0,10,size=(8,3))
B = np.random.randint(0,10,size=(2,2))

print("Original Array A:", A)
print("Original Array B:", B)
B0 = set(B[0])
B1 = set(B[1])

mask1 = np.isin(A,list(B0)).any(axis=1)
mask2 = np.isin(A,list(B1)).any(axis=1)

combined_mask = np.logical_and(mask1,mask2)

print("Selected Rows Of A:", A[combined_mask])
```

**5.** Дана 10x3 матрица, найти строки из неравных значений (например строка [2,2,3] остается, строка [3,3,3] удаляется)

**5.** Given a 10x3 matrix, find the rows consisting of non-equal values (e.g., row [2,2,3] remains, row [3,3,3] is removed)

```python

A = np.random.randint(0, 2, size=(10, 3))
print("Original Array:", A)

first_column = A[:, 0:1]
is_equal_to_first = (A == first_column)
all_same = is_equal_to_first.all(axis=1)

mask = ~all_same
result = A[mask]
print("Result:",result)
```

rectangular matrix. Example: For X = np.array([[1, 0, 1], [2, 0, 2], [3, 0, 3], [4, 4, 4]]), the answer is 3.

```python
#Python Implementation
Product = 1
X = np.array([
    [1,0,1],
    [2,0,2],
    [3,0,3],
    [4,4,4]
])
for indexI, valI in enumerate(X):
      for indexJ, valJ in enumerate(valI):
          if indexI == indexJ:
              if X[indexI][indexJ] != 0:
                  Product *= X[indexI][indexJ]
print(Product)

```

 * __Task 3__: Find the maximum element in vector x among elements preceded by a zero. Example: For x = np.array([6, 2, 0, 3, 0, 0, 5, 7, 0]), the answer is 5.

```python
#Python Implementation
x = np.array([6,2,0,3,0,0,5,7,0])

max_element = 0
for index, val in enumerate(x[1:],start=1):
  if x[index-1] == 0:
    if val > max_element:
      max_element = val
    else:
      continue
print(max_element)
```

* __Задача 4__: Реализовать кодирование длин серий (Run-length encoding). Для некоторого вектора x необходимо вернуть кортеж из двух векторов одинаковой длины. Первый содержит числа, а второй - сколько раз их нужно повторить.  
 Например, для x = np.array([2, 2, 2, 3, 3, 3, 5]) ответ (np.array([2, 3, 5]), np.array([3, 3, 1])).

 * __Task 4__: Implement Run-length encoding (RLE). For a vector x, return a tuple of two arrays of equal length: the first contains the values, and the second contains their counts. Example: For x = np.array([2, 2, 2, 3, 3, 3, 5]), the answer is (np.array([2, 3, 5]), np.array([3, 3, 1])).

```python
#Python Implementation
def rle_python(x):
    if len(x) == 0:
        return np.array([]), np.array([])

    values = []
    counts = []

    current_val = x[0]
    count = 0

    for val in x:
        if val == current_val:
            count += 1
        else:
            values.append(current_val)
            counts.append(count)
            current_val = val
            count = 1

    values.append(current_val)
    counts.append(count)

    return np.array(values), np.array(counts)

vals, counts = rle_python(x)
print(vals, counts)
```

```python
#Numpy Implementation
x = np.array([2,2,2,3,3,3,5])
def rle_numpy(x):
    if len(x) == 0:
        return np.array([]), np.array([])

    locs = np.where(x[1:] != x[:-1])[0]
    boundaries = np.concatenate(([ -1], locs, [len(x) - 1]))

    counts = np.diff(boundaries)
    values = x[boundaries[1:]]

    return values, counts

vals, counts = rle_numpy(x)
print(vals, counts)
```

* __Задача 5__: Даны две выборки объектов - X и Y. Вычислить матрицу евклидовых расстояний между объектами. Сравните с функцией scipy.spatial.distance.cdist по скорости работы.

* __Task 5__: Given two samples of objects - X and Y. Calculate the matrix of Euclidean distances between objects. Compare the performance with the scipy.spatial.distance.cdist function.

```python
#Python Implementation
import math
import time
from scipy.spatial.distance import cdist

def python_euclidean_dist(X, Y):
    num_x = len(X)
    num_y = len(Y)
    dist_matrix = [[0.0 for _ in range(num_y)] for _ in range(num_x)]

    for i in range(num_x):
        for j in range(num_y)
            squared_sum = 0.0
            for k in range(len(X[i])):
                squared_sum += (X[i][k] - Y[j][k])**2
            dist_matrix[i][j] = math.sqrt(squared_sum)
    return dist_matrix

#Numpy Implementation
def numpy_euclidean_dist(X, Y):
    x_2 = np.sum(X**2, axis=1, keepdims=True)
    y_2 = np.sum(Y**2, axis=1)

    xy = 2 * np.dot(X, Y.T)
    return np.sqrt(np.maximum(x_2 - xy + y_2, 0))


# Testing
N, M, D = 300, 300, 50
X_np = np.random.rand(N, D)
Y_np = np.random.rand(M, D)

# Convert to lists for the Python implementation test to be accuratee
X_list = X_np.tolist()
Y_list = Y_np.tolist()

print(f"Testing distance matrix calculation ({N}x{M})...\n")

# 1. Pure Python
start = time.time()
res_py = python_euclidean_dist(X_list, Y_list)
t_py = time.time() - start
print(f"Python Implementation: {t_py:.5f} seconds")

# 2. NumPy
start = time.time()
res_np = numpy_euclidean_dist(X_np, Y_np)
t_np = time.time() - start
print(f"NumPy Implementation:  {t_np:.5f} seconds")

# 3. SciPy cdist
start = time.time()
res_sp = cdist(X_np, Y_np, 'euclidean')
t_sp = time.time() - start
print(f"SciPy cdist:           {t_sp:.5f} seconds")
```
